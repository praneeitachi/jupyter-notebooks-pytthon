```python
# a simple example for a function
# req: 1. clear the screen ==> give a text ==> wait 2 sec
    #  2. printt the dir
import os
import time
print("please wait, we are clearing the screen")
time.sleep(2)
os.system("cls")
print("please wait, we are fetching the directories")
time.sleep(2)
os.system("dir")
```

```python
# platform independent
import os
import time
import platform
if platform.system=="Windows":
    print("please wait, we are clearing the screen")
    time.sleep(2)
    os.system("cls")
    print("please wait, we are fetching the directories")
    time.sleep(2)
    os.system("dir")
else:
    print("please wait, we are clearing the screen")
    time.sleep(3)
    os.system("clear")
    print("please wait, we are fetching the directories")
    time.sleep(3)
    os.system("ls-lrt")
```

```python
# in above code we are repeatin logic
# variables are 3 in that logic cls,sleep,dir and clear,sleep,ls-lrt
# reduce the code for better execution
# we can achieve this using functions
import os
import time
import platform
def my_code(x,y,z):
    print("please wait, we are clearing the screen")
    time.sleep(x)
    os.system(y)
    print("please wait, we are fetching the directories")
    time.sleep(x)
    os.system(z)
if platform.system()=="Windows":
    my_code(2,'cls','dir')
else:
    my_code(3,'clear','ls-lrt')
```
