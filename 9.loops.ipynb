### Range
------------------
==> range is used only on numbers
==> 5 ==> default it takes 0,1,2,3,4
==> range(start,stop,step/increment)
    start-- start of a range default 0
    stop --- stopping a range value
    step --- increase with a value 
==> to convert range into list
ex: list(range(5)
very useful

```python
# using range print values in a list
list=[1,2,6,8,"python"]
print(range(len(list)))
index=range(len(list))
for item in index:
    print(f"for index {item} ---> value is {list[item]}")
# we are using range function and replacing with index and calling values
```

```python
# in a list inside tuples
#2. unpacking a list
list=[(1,2),(3,4),(5,6)]
for each_item in list:
    print(each_item)
for x,y in list:
    print(x)
    print(y)
```

```python
# 3 handling dictionaries
# they are very important 
dict={1:"one",2:"two",3:"three"}
for item in dict:
    print(item)
print("-------------------------")
# by default it will print only keys 
# you can use dict.keys() also
for item in dict.values():
    print(item)
print("--------------------------")   
# we use dict.values() to print values
for item in dict.items():
    print(item)
print("--------------------------")
# we are using items() to call each item
for key,value in dict.items():
    print(f"key-->{key} & value--->{value}")
print("--------------------------")
```

## While loop
if the condition is true it will run till we stop
no limitation ===> continously it will run
to repeat a logic infinite number of times 

```python
# in reak time you dont know how many times you havee to run a loop
while False:
    print("sample while loop") 
# we are stopping this it will run infinite times 
# instead of True ===> False we will use
```

```python
# 2 using time module to run every second if while is true
import time
while False:
    print("we are monitoring for every second")
    time.sleep(1)
```

```python
# we can give a value and print it infinite times also
value=4
while False:
    print(value)
    value=value+4
```

```python
# we can limit while loop to execute also
cnt=1
while cnt<=6:
    print("hello")
    cnt=cnt+1
```

## Loop Control Statements

==> for loop --- we know how many times we can iterate
==> while loop --- we dont know how many times so we will use this 
==> we can control the loop using
    1. break --- break the loop
    2. continue -- except the logic under continue it will run all
    3. pass 

```python
# using for loop
for each in [1,2,3,4,5]:
    print(each)
```

```python
# to brak the loop without executing it
# we can use brak
for each in [1,2,3,4,5]:
    print(each)
    break
print("loop is stopped")
```

```python
 # we can brak the loop at whatever you want 
for each in [1,2,3,4,5]:
    print(each)
    if each==4:
        break
print("loop is stopped after we attained a desired value")
```

```python
#continue
# if we give this statement it wont run that logic for that condition
list=[1,2,3,4,5,2,7,5]
for x in list:
    if x==7:
        continue
        print("7 is skipped")
    print(x)
```
