```python
# req: we want is and it on the above string
# create a pattern for the requirement
import re
string="python is easy but it requires to practise"
pattern="i[st]"
#pattern="i[ts]" both are same
print(re.findall(pattern,string))
print(len(re.findall(pattern,string)))
```

```python
import re
string="python is easy but it requires to @practise 101 rues 103"
pattern1="a"
pattern2="[st]"
# searches s and t chars in a string
pattern3="[a-de-g]" # ==> a,b,c,d,e,f,g
#pattern="i[ts]" both are same
pattern4="\w"
# prints these chaars a-z,A-Z,0-9,_ 
pattern5="\w\w\w"
# we can give these for any \ symbols \W\W or \d\d\d etc
pattern6="\W"
# all the chars except \w
pattern7="\d"
# prints all numbers in a string
pattern8="\D"
# except numbers it will print all
print(re.findall(pattern1,string))
print(re.findall(pattern2,string))
print(re.findall(pattern3,string))
print(re.findall(pattern4,string))
print(re.findall(pattern5,string))
print(re.findall(pattern6,string))
print(re.findall(pattern7,string))
print(re.findall(pattern8,string))
```

```python
# we can pass these to our previous patterns
import re
string="python is easy but it requires to practise"
pattern="p"
pattern1="^p"
# prints the first character in a first line if p is there prints
pattern2="y"
pattern3="^y"
#here there is no y at start 
print(re.findall(pattern,string))
print(re.findall(pattern1,string))
print(re.findall(pattern2,string))
print(re.findall(pattern3,string))
```

```python
import re
string="python is easy to learn but it requires to practise nd learn"
pattern="e"
pattern1="e$"
# prints the first character in a first line if p is there prints
pattern2="learn"
pattern3="learn$"
#here there is no y at start 
print(re.findall(pattern,string))
print(re.findall(pattern1,string))
print(re.findall(pattern2,string))
print(re.findall(pattern3,string))
```

```python
# for space we can use \b ==> \\b 
import re
string="python is easy to learn but it requireslearning to practise and learn"
# word  ==> two sides space will be there
# starting word and ending word by default has space 
pattern="\\blearn"
print(re.findall(pattern,string))
# here left side space we provided
pattern="\\blearn\\b"
print(re.findall(pattern,string))
# both sides we are providing
pattern="\\blearn\\b$"
print(re.findall(pattern,string))
# you can combine with ^ or $ also 
pattern="\\Blearn\\B"
print(re.findall(pattern,string))
# it prints from requires'learn'ing this string
# instaed of using \\ we also can use raw string "r"
pattern=r"\Blearn\B"
print(re.findall(pattern,string))
```

```python
import re
string="pyhon is aaaa aa aaa  aa aa aa aaaaa"
pattern=r"\ba{2}\b"
# repeat a ==> 2
print(re.findall(pattern,string))
pattern="a{2,5}"
# a==> 2 a==>3 a==>4 repeated print
print(re.findall(pattern,string))
pattern="a{2,}"
# a==> 2 > i,e two or more 
print(re.findall(pattern,string))
```

```python
# between two letters
import re
string="pyhon is xaaaax xaax xaaax xaax xaax xaax xaaaaax xx"
pattern=r"\bxa+x\b"
# + one or more not all 
print(re.findall(pattern,string))
pattern=r"\bxa*x\b"
# * 0 or more, it prints xx also 
print(re.findall(pattern,string))
pattern=r"\bxa?x\b"
# a ==> one or none 
print(re.findall(pattern,string))
```

```python
# we will use flags sometimes based on our requirement
import re
string="This is this but not that THIS, THIs is THis"
pattern=r"this"
print(re.findall(pattern,string))
# you want to print all "this "==> ignore case
print(re.findall(pattern,string,re.I))
```
