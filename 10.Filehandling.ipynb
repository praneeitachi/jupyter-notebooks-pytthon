# Working With Text Files
-------------------------------------------------------------------------
1. operations you can perform for fileas are read,write,append
2. we can create a file with any of the operations 
3. we will use open() and close() functions to open a file and close it

```python
# simple creation of a empty text file
file=open("new.txt","w")
print(dir(file))
file.close()
# these are the operations we can perform on a file 
```

```python
# some operations you can do are 
file=open("new.txt","w")
print(file.mode)
print(file.readable())
print(file.writable())
file.close()
```

```python
# 1st way ---> adding lines into a created file
file=open("new.txt","w")
# if content is there use "a" else you can use "w" also
file.write("this is a first line \n")
file.write("this is a second line \n")
# here we are creating lines one by one insead we can create lines and pass them
file.close()
# 2nd way ----> adding multipple lines
content=["this is a 1stline \n","this is a 2nd line \n"]
file=open("new.txt","w")
file.writelines(content)
file.close()
#3rd way --> using loops to add content ---> preferred way 
content=["this is a 1stline","this is a 2nd line","this is a 3rd line"]
file=open("new.txt","w")
for x in content:
    file.write(x+"\n")    
file.close()
```

```python
# reading a created file
file=open("new.txt","r")
read=file.read()
print(read)
print(type(read))
```

```python
# its a string , we want to read content as a list 
file=open("new.txt","r")
read=file.readlines()
print(read)
print(type(read))
```

```python
# reading 2 lines inthat list (last 2 or first 2)
file=open("new.txt","r")
read=file.readlines()
for x in range(len(read)):
    print(read[x])
    
```

```python
# its printing directly, its a string
# 1st wat --> we want as a tabular data or list we want
import csv
input="C:\\Users\\prane\\Desktop\\Python\\Codes\\emp.csv"
csv=open(input,"r")
content=csv.readlines()
for x in content:
    print(x.split("\n"))
print(type(content))
```

```python
import csv
input="C:\\Users\\prane\\Desktop\\Python\\Codes\\emp.csv"
file=open(input,"r")
content=csv.reader(file,delimiter=",")
for x in content:
    print(x)
file.close
print(type(content))
'''
if reader is not avilable
print(dir(csv)) ==> see if reader is there if not
pip install csv-reader
'''
# if csv using pipe is there use delimiter " |"
```

```python
# to print a header we can easily do usong list(content)[0] in one way 
# we can also use next()--> next means the cursor goes to next line 
import csv
input="C:\\Users\\prane\\Desktop\\Python\\Codes\\emp.csv"
file=open(input,"r")
content=csv.reader(file,delimiter=",")
header=next(content)
print(f"the header is {header}")
print(len(list(content)))
print(f"number of rows excluding header are {len(list(content))}")
file.close
```
