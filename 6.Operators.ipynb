## Identity & Membership Operators

```python
a=type(2)
b=type("string")
print(a is b)
print(a is not b)
```

```python
java_ver=["java1.1","java1.2","java1.3"]
java_host="java1.1"
print(java_host in java_ver )
java_host="java1.5"
print(java_host not in java_ver )
```

```python
# simple if satement we can define using this as
java_ver=["java1.1","java1.2","java1.3"]
java_host="java1.1"
if java_host in java_ver:
    print(f"you can run the application using {java_host}")
else:
    print(f"you cant run the application using {java_host}")
```

## Logical Operators

```python
# and condition can be represented using all also 
print(1<2 and 2<3 and 3<4)
all([1<2,2<3,3<4])
```

```python
# or can be represented with any function also
print(2>3 or 3>4 or "a"<"b")
any([2>3,3>4,4<5])
```
