```python
# it will check all excepts if error matches with anyone it executs
# if it doesnt match it goes to Exception as e
try:
    print(a)
except NameError:
    print("variable is not defined") 
except Exception as e:
    print(e) 
```

```python
# finally block
# it executes irrespective of try executes or error executes just like a print
#at ebd it will execute 
try:
    print(a)
except NameError:
    print("variable is not defined") 
except Exception as e:
    print(e) 
finally:
    print("you have to define a variable in order to resolve this error")
```

```python
try:
    a=10
    print(a)
except NameError:
    print("variable is not defined") 
except Exception as e:
    print(e) 
finally:
    print("you have to define a variable in order to resolve this error")
```

```python
# try, else block 
# if try executes ==> else executes
# if try not execute ==> else not execute 
try:
    a=10
    print(a)
except NameError:
    print("variable is not defined") 
except Exception as e:
    print("the error is",e) 
else:
    print("if a is defined , no exception, good code")
finally:
    print("you have to define a variable in order to resolve this error")
```
